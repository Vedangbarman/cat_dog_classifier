```python
import pandas as pd
import numpy as np

import matplotlib as mt
import seaborn as sns
```

```python
from sklearn.datasets import load_breast_cancer, load_digits

# Load the breast cancer numeric dataset
cancer_data = load_breast_cancer()
cancer_data['target'] = cancer_data.target
```

```python
data = pd.DataFrame(cancer_data.data, columns = cancer_data.feature_names)
```

```python
data['target'] = cancer_data.target
```

```python
pd.set_option('display.max_columns', None)
```

```python
data.head()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = data.drop('target',axis = 1)
y = data.target
```

```python
scaler.fit(X)
```

```python
data_log = np.log1p(X)
data_scaled_array=scaler.fit_transform(data_log)
data_scaled = pd.DataFrame(data_scaled_array, columns=X.columns)
```

```python
data_scaled.describe()
```

```python
data.describe()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = data_scaled
y = data.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
import lazypredict
from lazypredict.Supervised import LazyClassifier
clf = LazyClassifier(verbose=1, ignore_warnings=False, custom_metric=None)
models, predictions = clf.fit(X_train, X_test, y_train, y_test)
display(models)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix,precision_score, recall_score,f1_score,roc_curve, roc_auc_score
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f"accuracy: {round(accuracy,2)}")
```

```python
from sklearn.metrics import mean_squared_error
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Generate the matrix
cm = confusion_matrix(y_test, y_pred)

# Plot it using Seaborn for a nice visual
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted Label')
plt.ylabel('Actual Label')
plt.title('Confusion Matrix')
plt.show()
```
